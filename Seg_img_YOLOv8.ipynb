```python
# Criando fun√ß√£o para mostrar a imagem
def mostrar(img):
  fig = plt.gcf() # gcf limpa os eixos
  fig.set_size_inches(16,10) # 16/10 tamanho da imagem em polegadas
  plt.axis('off') # off para n√£o mostrar os valores nos eixos x e y
  plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)) # Convertendo para o formato padr√£o RGB
  plt.show()
```

```python
seg = cv2.imread('/content/runs/segment/predict/avi√£o.jpg')
mostrar(seg)
```
