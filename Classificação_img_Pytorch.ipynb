```python
# Importando Bibliotecas
import torch
from torch import nn, optim
from torchvision import datasets, transforms
torch.__version__
```

```python
# Função para manter o controle e gerar os mesmos valores
torch.manual_seed(123)
```

```python
# Construindo o modelo
classificador = nn.Sequential(nn.Conv2d(in_channels=3, out_channels=32, kernel_size = 3), # camada de convuloção, 3 canais, 32 filtros, kernel_size matriz 3x3
                              nn.ReLU(),
                              nn.BatchNorm2d(num_features=32),  # batchnorm para normalizar as camadas ocultas
                              # (64 - 3 + 1) / 1 = 62x62       # tamanho da imagem depois da primeira camada de convolução
                              nn.MaxPool2d(kernel_size = 2),
                              # 31x31  # depois do maxpool a imagem reduz pela metade
                              nn.Conv2d(32, 32, 3), # input, filtros e kernel
                              nn.ReLU(),
                              nn.BatchNorm2d(32),
                              # (31 - 3 + 1) / 1 = 29x29 após a segunda camada a imagem reduz para 29x29
                              nn.MaxPool2d(2), # filtro (matriz) 2x2
                              #  14x14 novamente caiu pela metade
                              nn.Flatten(),
                              # 6272 -> 128 -> 128 -> 1
                              nn.Linear(in_features=14*14*32, out_features=128), # neurônios na camada de entreda 14*14*32, 128 na camada oculta
                              nn.ReLU(),
                              nn.Dropout(0.2), # 20% dos neurônios serão zerados
                              nn.Linear(128, 128), # 128 neuronios da camada anterior + 128 da nova camada
                              nn.ReLU(),
                              nn.Dropout(0.2),
                              nn.Linear(128, 1), # 1 neurônio na saída
                              nn.Sigmoid())
```

```python
# Definindo o otimizador
criterion = nn.BCELoss()
optimizer = optim.Adam(classificador.parameters())
```

```python
# Carregando a base de dados de treino e teste
treino = '/content/OIDv4_ToolKit/OID/Dataset/train'
teste = '/content/OIDv4_ToolKit/OID/Dataset/test'
```

```python
# Criando um transformador para a base de dados de treino
transform_treino = transforms.Compose(
    [
     transforms.Resize([64, 64]), # transformando as dimensões para 64X64 para padronizar
     transforms.RandomHorizontalFlip(), # transformação de alinhamento horizontal da imagem
     transforms.RandomAffine(degrees=7, translate=(0, 0.07), shear=0.2, scale=(1, 1.2)), # rotação na imagem, scale para zoom
     transforms.ToTensor() # tranforma no formato tensor
    ]
)
```

```python
# Criando transformador para a base de dados de teste
transform_teste = transforms.Compose(
    [
     transforms.Resize([64, 64]), # Redimensionamento
     transforms.ToTensor()         # Convertendo para tensor
    ]
)
```

```python
# Criando a base de dados de treino. Imagefolder para ler as imagens
treino_dataset = datasets.ImageFolder(treino, transform=transform_treino)
```

```python
# Observando
treino_dataset
```

```python
# Fazendo a normalização e será a base para fazer o treinamento. Shuffle para misturar os dados
treino_loader = torch.utils.data.DataLoader(treino_dataset, batch_size = 32, shuffle=True)
```

```python
# Criando a base de dados de teste. Imagefolder para ler as imagens
teste_dataset = datasets.ImageFolder(teste, transform=transform_teste)
```

```python
# Observando
teste_dataset
```

```python
# Normalizando
teste_loader = torch.utils.data.DataLoader(teste_dataset, batch_size = 32, shuffle=True)
```

```python
# Observando a conexão com a GPU
device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
device
```

```python
# Classificador do dispositivo para ser executado na GPU
classificador.to(device)
```

```python
# Definindo função para treinar o modelo
def training_loop(loader, epoch): # Definindo função que recebe a base de dados e as épocas
    running_loss = 0.             # taxa de erro
    running_accuracy = 0.         # taxa de acerto

    for i, data in enumerate(loader): # for para percorrer a base de dados
        inputs, labels = data         # buscando os pixels e as respostas
        inputs, labels = inputs.to(device), labels.to(device) # enviando para a GPU

        optimizer.zero_grad()           # zerando o gradiente
        outputs = classificador(inputs) # variável de retorno da classificação

        loss = criterion(outputs, labels.float().view(*outputs.shape)) # calculando erro e convertendo para float
        loss.backward()   # Inicializando o backpropagation

        optimizer.step()  # Atualizando os pesos

        running_loss += loss.item()

        predicted = torch.tensor([1 if output > 0.5 else 0 for output in outputs]).to(device)

        equals = predicted == labels.view(*predicted.shape) # verificando se o valor previsto é igual ao do label

        accuracy = torch.mean(equals.float()) # calculando accuracy e tiramos a média
        running_accuracy += accuracy          # Atualizando o running accuracy

        # Imprimindo os dados referentes a esse loop
        print('\rÉPOCA {:3d} - Loop {:3d} de {:3d}: perda {:03.2f} - precisão {:03.2f}'.format(epoch + 1, i + 1, len(loader), loss, accuracy), end = '\r')

    # Imprimindo os dados referentes a essa época
    print('\rÉPOCA {:3d} FINALIZADA: perda {:.5f} - precisão {:.5f}'.format(epoch + 1, running_loss/len(loader),
                    running_accuracy/len(loader)))
```

```python
for epoch in range(15): # Treinando por 15 épocas
    print('Treinando...')
    training_loop(treino_loader, epoch) # Passando a base de dados de treino e atualizar os pesos
    classificador.eval() # Classificador em modo de avaliação para ter accuracy durante o treinamento
    print('Validando...')
    training_loop(teste_loader, epoch) # Fazendo o teste época por época
    classificador.train() # Classificador em modo de treino para atualizar os pesos
```

```python
# Função para classificar imagens
def classificar_imagem(fname):
  from PIL import Image  # Abrir imagens do python
  import matplotlib.pyplot as plt
  imagem_teste = Image.open(teste + '/' + fname) # leitura da imagem de teste
  plt.imshow(imagem_teste)
  # pré processamento da imagem de teste para passar para a rede neural
  import numpy as np
  imagem_teste = imagem_teste.resize((64, 64))
  imagem_teste = np.array(imagem_teste.getdata()).reshape(*imagem_teste.size, 3) # convertendo para o numpy
  imagem_teste = imagem_teste / 255   # fazendo a normalização
  imagem_teste = imagem_teste.transpose(2, 0, 1)
  imagem_teste = torch.tensor(imagem_teste, dtype=torch.float).view(-1, *imagem_teste.shape)

  classificador.eval() # classificador em mode de avaliação
  imagem_teste = imagem_teste.to(device) # mandando a imagem para a GPU
  output = classificador.forward(imagem_teste) # saída
  if output > 0.5:  # se a saída for maior que 0.5 mostrará 1
    output = 1
  else:
    output = 0
  print('Previsão: ', output)

  idx_to_class = {value: key for key, value in teste_dataset.class_to_idx.items()} # Indicando se é carro ou caminhão

  return idx_to_class[output]
```
